```python
string = "ProgramlamaÖdeviİleriSeviyeVeriYapılarıveObjeleripynb"

# Boş bir sözlük oluşturun
frekans = dict()

# String içindeki her karakteri döngü ile kontrol edin
for char in string:
    # Karakter zaten frekans sözlüğünde mevcutsa, değerini 1 artırın
    if char in frekans:
        frekans[char] += 1
    # Karakter daha önce hiç görülmediyse, sözlüğe ekleyin ve değerini 1 yapın
    else:
        frekans[char] = 1

# Frekansları yazdırın
for char, count in frekans.items():
    print(char, ":", count)
```

```python
isim = ["Kerim", "Tarık", "Ezgi", "Kemal", "İlkay", "Şükran", "Merve"]
soyisim = ["Yılmaz", "Öztürk", "Dağdeviren", "Atatürk", "Dikmen", "Kaya", "Polat"]

# "isim" ve "soyisim" listelerini birleştirin
isim_soyisim = list(zip(isim, soyisim))

# "isim"e göre sıralı bir şekilde ekrana yazdırın
for ad, soyad in sorted(isim_soyisim):
    print(ad, soyad)
```

```python
def asal_sayi_bul(sayi):
    asal_sayilar = []
    for i in range(2, sayi+1):
        asal = True
        for j in range(2, i):
            if i % j == 0:
                asal = False
                break
        if asal:
            asal_sayilar.append(i)
    return asal_sayilar

sayi = int(input("Asal sayıları bulmak için bir sayı girin: "))
asal_sayilar = asal_sayi_bul(sayi)
print("1 ile {} arasındaki asal sayılar: {}".format(sayi, asal_sayilar))
```
