```python
class Yazılımcı():
    
    def __init__(self,isim,soyisim,calistigiyer,maas,diller,kitapsayisi):
        self.isim = isim
        self.soyisim = soyisim
        self.calistigiyer = calistigiyer
        self.maas = maas
        self.diller = diller
        self.kitapsayisi = kitapsayisi
        
    def bilgilerigoster(self):
        print("""
        Yazılımcının özellikleri:
        
        İsim: {}
        
        Soyisim: {}
        
        Çalıştığı yer: {}
        
        Maaşı: {}
        
        Diller: {}
        
        Okuduğu Kitap sayısı: {}
        
        """.format(self.isim,self.soyisim,self.calistigiyer,self.maas,self.diller,self.kitapsayisi))
    
    def zam_yap(self,zam_miktari):
        print("zam yapıldı, hayırlı olsun")
        self.maas += zam_miktari
    
    def dil_ekle(self,yeni_dil):
        print("yeni diliniz eklendi")
        self.diller.append(yeni_dil)
    
    def kitapsayisiekle(self,kitap_ekle):
        print("tebrikler kütüphaneniz güncellendi.")
        self.kitapsayisi += kitap_ekle
        
        
        
        
    
```

```python
yazılımcı1 = Yazılımcı("Ahmet Yavuz","Çakmas","Silikon Vadisi",12000,["python"],23)
```

```python
yazılımcı1.bilgilerigoster()
```

```python
yazılımcı1.kitapsayisiekle(22)
```

```python
yazılımcı1.zam_yap(15000)
```

```python
yazılımcı1.bilgilerigoster()
```

```python
yazılımcı1.kitapsayisiekle(-12)
```

```python
yazılımcı1.bilgilerigoster()
```
