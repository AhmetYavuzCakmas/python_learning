```python
def mukemmelsayibulma(sayi):
    bolen_toplam = 0
    for i in range(1, sayi):
        if sayi % i == 0:
            bolen_toplam += i
    return bolen_toplam == sayi

for i in range(1, 1001):
    if mukemmelsayibulma(i):
        print(i)
```

```python
def find_gcd(num1, num2):
    small = min(num1, num2)
    big = max(num1, num2)
    
    for i in range(small, 0, -1):
        if big % i == 0 and small % i == 0:
            return i
def find_lcm(num1, num2):
    gcd = find_gcd(num1, num2)
    return (num1 * num2) // gcd
num1 = int(input("İlk sayıyı girin: "))
num2 = int(input("İkinci sayıyı girin: "))

lcm = find_lcm(num1, num2)
print("En küçük ortak kat:", lcm)
```

```python
def okunus(sayi):
    birler = ["", "Bir", "İki", "Üç", "Dört", "Beş", "Altı", "Yedi", "Sekiz", "Dokuz"]
    onlar = ["", "On", "Yirmi", "Otuz", "Kırk", "Elli", "Altmış", "Yetmiş", "Seksen", "Doksan"]

    if sayi < 10 or sayi > 99:
        return "Lütfen 2 basamaklı bir sayı girin!"

    onlar_basamagi = sayi // 10
    birler_basamagi = sayi % 10

    return onlar[onlar_basamagi] + " " + birler[birler_basamagi]

sayi = int(input("2 basamaklı bir sayı girin: "))

print(okunus(sayi))
```
