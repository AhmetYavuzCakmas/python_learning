```python
def asal_mi(sayi):
    if (sayi == 1):
        return False
    elif (sayi == 2):
        return True
    else:
        for i in range(2,sayi):
            if (sayi % i == 0):
                return False
        return True
    
    
while True:
    sayi = input("sayı: ")
    if (sayi == "q"):
        break
    else:
        sayi = int(sayi)
        if (asal_mi(sayi)):
            print(sayi, "asal sayıdır")
        else:
            print(sayi, "asal sayı değildir")
```

```python
def tambolenleribulma(sayi):
    tam_bolenler =[1]
    for i in range(2,sayi + 1):
            if ( sayi % i == 0):
                tam_bolenler.append(i)
    return tam_bolenler

while True:
    sayi = input("sayı: ")
    if (sayi == "q"):
        print("programdan sonlandırılıyor..")
        break
    else:
        sayi = int(sayi)
        print("tam bolenler",tambolenleribulma(sayi))
```

```python
def mukemmelsayibulma(sayi):
    tam_bolenler =[1]
    sayi = int(sayi)
    for i in range(2,sayi):
            if ( sayi % i == 0):
                tam_bolenler.append(i)
    toplam = sum(tam_bolenler)
    return toplam

while True:
    sayi = input("sayı: ")
    if (sayi == "q"):
        print("program sonlandırılıyor..")
        break
    else:
        sayi = int(sayi)
        if sayi == mukemmelsayibulma(sayi):
            print("toplamı",mukemmelsayibulma(sayi))
            print("{} mükemmel sayıdır".format(sayi))
            
        else:
            sayi = int(sayi)
            print("toplamı",mukemmelsayibulma(sayi))
            print("{} mükemmel sayı değildir".format(sayi))
```

```python
birler  =   [ "" , "Bir" , "İki" , "Üç" , "Dört" , "Beş" , "Altı" , "Yedi" , "Sekiz" , "Dokuz" ] 
onlar  =  [ "" , "On " , "Yirmi" , "Otuz" , "Kırk" , "Elli" , "Altmış" , "Yetmiş" , "Seksen" ,"Doksan" ]

def  okunus( sayı ): 
    birinci  =  sayı  %  10 
    ikinci  =  sayı  //  10
    
    return  onlar[ ikinci ]  +  " "  +  birler [ birinci ]

    
sayı  =   int ( input ( "Sayı:" ))

print ( okunus ( sayı ))
```
